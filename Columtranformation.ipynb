```python
traffic_cat=['Jam', 'High', 'Medium', 'Low']
#weather_cat=['Fog', 'Stormy', 'Sandstorms', 'Windy', 'Cloudy', 'Sunny']
festivel_cat=['Yes','No']

```

```python
def ohe_many_cat(data, dataset):
    for i in data:
        top_10 = [x for x in dataset[i].value_counts().sort_values(ascending=False).head(10).index]
        for label in top_10:
            dataset[label] = np.where(dataset[i] == label, 1, 0)
        dataset[[i]+top_10].head(20)
        
```

```python
from sklearn.impute import SimpleImputer ##for handling the missing values of the data
from sklearn.preprocessing import StandardScaler ##handling feature sacling
from sklearn.preprocessing import OrdinalEncoder ## for ordinal encoding
from sklearn.preprocessing import OneHotEncoder  ## for OneHot Encoding
#for creating pipelines 
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer 
```

```python
##
num_pipeline=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
)
```

```python
cat_pipeline_one=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy="most_frequent")),
        ('ordinalencoder',OrdinalEncoder(categories=[traffic_cat,festivel_cat])),
        ('scaler',StandardScaler())
    ]
)
```

```python
cat_pipeline_ohe=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy="most_frequent")),
        ('onehotencoder',OneHotEncoder(sparse=False,drop='first')),
        ('scaler',StandardScaler())
        
    ]

    
)
```

```python
##model traning
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
## Train multiple models

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)

    #Make Predictions
    y_pred=model.predict(x_test)

    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
